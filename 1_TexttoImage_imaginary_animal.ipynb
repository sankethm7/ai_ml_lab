```python
from diffusers import DiffusionPipeline
import torch

model_id = "CompVis/stable-diffusion-v1-4"
#model_id = "stabilityai/stable-diffusion-2-1"
device = "cuda"
#pipeline = DiffusionPipeline.from_pretrained("black-forest-labs/FLUX.1-dev", torch_dtype=torch.float16)
pipeline = DiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)


pipeline.to(device)
prompt = "An image of animal having cow like face, giraffe like neck, lion like body, horse like legs and elephant like tail"
image = pipeline(prompt).images[0]

image.save("imaginary anmimal using stable-diffusion-2-1.png")

```
